```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
data = pd.read_csv('train.csv')

# Display the first few rows of the dataset
print(data.head())

# Summary statistics of the numerical columns
print(data.describe())

# Count of passengers in each class
class_counts = data['pclass'].value_counts()
print(class_counts)

# Count of passengers who survived and who didn't
survival_counts = data['survived'].value_counts()
print(survival_counts)
```

```python
# Distribution of ages
plt.figure(figsize=(8, 6))
sns.histplot(data['age'].dropna(), kde=True)
plt.title('age distribution')
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

```python
# Survival rate based on gender
plt.figure(figsize=(6, 6))
sns.barplot(x='sex', y='survived', data=data, errorbar=None)
plt.title('Survival Rate based on Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Survival rate based on passenger class
plt.figure(figsize=(6, 6))
sns.barplot(x='pclass', y='survived', data=data, errorbar=None)
plt.title('Survival Rate based on Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
numeric_columns = data.select_dtypes(include=[float, int]).columns
correlation_matrix = data[numeric_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
